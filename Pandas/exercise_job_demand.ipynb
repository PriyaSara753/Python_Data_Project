```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"F:\Jupyter\data_jobs.csv")
df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
```

```python
df_usa = df.copy()
```

```python
id(df)==id(df_usa)
```

```python
df_usa["job_posted_month"]=df_usa['job_posted_date'].dt.strftime("%B")
df_usa.head(5)
```

```python
df_usa_pivot=df_usa.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_usa_pivot
```

```python
df_usa_pivot.reset_index(inplace=True)
```

```python
df_usa_pivot["month_no"]=pd.to_datetime(df_usa_pivot['job_posted_month'],format="%B").dt.month
df_usa_pivot
```

```python
df_usa_pivot.sort_values('month_no',inplace=True)
df_usa_pivot
```

```python
df_usa_pivot.set_index('job_posted_month',inplace=True)
df_usa_pivot
```

```python
df_usa_pivot.drop(columns='month_no',inplace=True)
```

```python
df_usa_pivot
```

```python
df_usa_pivot.plot(kind='line')
```

```python
top_3 = df['job_title_short'].value_counts().head(3)
top_3 =top_3.index.tolist()
```

```python
df_usa_pivot[top_3].plot(kind='line')
plt.title("Monthly job postings for top data jobs in US")
plt.xlabel('2023')
plt.ylabel('Job count')
plt.legend()
plt.show()
```

```python
df_us_software_pivot=pd.read_csv("F:\Jupyter\software_jobs.csv",index_col='job_posted_month')
df_us_software_pivot
```

```python
df_usa_merged= df_usa_pivot.merge(df_us_software_pivot,on='job_posted_month')
df_usa_merged
```

```python
top_5 = df_usa_merged.sum().sort_values(ascending=False).head(5).index.tolist()
top_5
```

```python
df_usa_merged[top_5].plot(kind='line')
plt.title("Monthly job postings for top Tech jobs in US")
plt.xlabel('2023')
plt.ylabel('Job count')
plt.legend()
plt.show()
```
