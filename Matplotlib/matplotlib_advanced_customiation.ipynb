```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
df=pd.read_csv(r"F:\Jupyter\data_jobs.csv")
df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
```

```python
df_DA_exploded = df_DA.explode('job_skills')
df_DA_pivot=df_DA_exploded.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)

df_DA_pivot.loc['Total']=df_DA_pivot.sum()
```

```python
df_DA_pivot
```

```python
df_DA_pivot=df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_pivot=df_DA_pivot.drop('Total',axis=0)
df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot
```

```python
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x,format='%m').strftime('%B'))
```

```python
df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
```

```python
df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')
df_DA_pivot
```

```python
df_DA_pivot
```

```python
df_DA_pivot.iloc[:,:5].plot(kind='line')
plt.title('TOP 5 skills for Data Analyst per monthly postings')
plt.xlabel('Count')
plt.ylabel('')
plt.show()
```

```python
df_DA_pivot.iloc[:,:5].plot(
    kind='line',
    linewidth=4,
    linestyle=':', # options -> '-', '--' , '-.' , ':'
    colormap='viridis', # options -> 'plasma' , 'inferno' , 'magma' , 'cividis'
    marker='o', # options -> 'o','#','+','*,'$,',',','.','1','2','3','4',...
    markersize=5,
    figsize=(10,5))


plt.title('TOP 5 skills for Data Analyst per month')
plt.xlabel('Count')
plt.ylabel('')
plt.show()
```
